```python
import pandas as pd

df = pd.read_csv("../data/raw/cicids2017_friday_ddos.csv")
```

```python
df = df.drop_duplicates()
df
```

Duplicate handling
- Exact duplicate flow records were removed to avoid repeated observations affecting later analysis. Such duplicates are common in flow-based network datasets and usually result from logging and aggregation processes.

```python
df = df[df[" Flow Duration"]> 0] & df[df[" Total Fwd Packets"] > 0]
df.describe()
```

Invalid / Unrealistic Values
- Flow records with non-positive durations were removed as they represent logging artefacts rather than valid network behaviour.

```python
df.isna().sum().sum()
```

Missing values
- After cleaning, no missing values remain in the dataset.

Data types
- Feature data types were verified to ensure numerical attributes are stored correctly, with non-numeric columns limited to categorical labels only.
