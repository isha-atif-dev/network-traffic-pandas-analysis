Label / Target Inspection

The class distribution shows that the dataset contains two traffic classes: DDoS and BENIGN.
There are 128,027 DDoS flows and 97,718 benign flows, indicating a moderate class imbalance with attack traffic slightly dominating the dataset.

This imbalance reflects traffic captured during an active attack scenario and suggests that evaluation metrics such as precision, recall, and F1-score may be more informative than accuracy alone in later machine learning stages.
